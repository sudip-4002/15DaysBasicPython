Project 1: Student Management System


```python
# Creating Student Data

students = []
```

```python
# Adding Student Data

name = input("Enter Student Name: ")
marks = int(input("Enter Student Marks: "))

student = {
    "name": name,
    "marks": marks
}

students.append(student)

print(students)
```

```python
# Displaying Student Records

for student in students:
    print(student)
```

```python
# Finding Average Marks

total = 0

for student in students:
    total += student["marks"]

average = total / len(students)

print("Average Marks:", average)
```

```python
# Finding Top Student

top_marks = 0
top_student = ""

for student in students:

    if student["marks"] > top_marks:
        top_marks = student["marks"]
        top_student = student["name"]

print("Top Student:", top_student)
print("Marks:", top_marks)
```

Project 2: ATM System


```python
# Creating ATM Balance

balance = 5000
```

```python
# Deposit Money

amount = int(input("Enter Deposit Amount: "))

balance += amount

print("Updated Balance:", balance)
```

```python
# Withdraw Money

amount = int(input("Enter Withdraw Amount: "))

if amount <= balance:
    balance -= amount
    print("Withdraw Successful")

else:
    print("Insufficient Balance")

print("Current Balance:", balance)
```

```python
# Checking Balance

print("Available Balance:", balance)
```

Project 7: Result Management System

```python
# Student Result Data

student = {
    "name": "Sudip",
    "marks": [70, 80, 90]
}

print(student)
```

```python
# Calculating Result

average = sum(student["marks"]) / len(student["marks"])

print("Average:", average)

if average >= 80:
    print("Grade: A")

elif average >= 60:
    print("Grade: B")

else:
    print("Grade: C")
```

Project 10: Portfolio Level Mini Project

```python
# Creating Student Records

students = []
```

```python
# Adding Multiple Students

for i in range(3):

    name = input("Enter Student Name: ")
    marks = int(input("Enter Student Marks: "))

    data = {
        "name": name,
        "marks": marks
    }

    students.append(data)

print(students)
```

```python
# Final Student Report System

highest_marks = 0
top_student = ""

total = 0

for student in students:

    total += student["marks"]

    if student["marks"] > highest_marks:
        highest_marks = student["marks"]
        top_student = student["name"]

average = total / len(students)

print("Average Marks:", average)
print("Top Student:", top_student)
print("Highest Marks:", highest_marks)
```
