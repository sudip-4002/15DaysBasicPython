```python
# Basic Function (No Arguments, No Return)
def greet():
    print("Hello, welcome to Python!")

greet()
```

```python
# Function with Argument
def greet_user(name):
    print("Hello", name)

greet_user("Sudip")
```

```python
# Function with Return Value
def add(a, b):
    return a + b

result = add(5, 3)
print(result)
```

```python
# Difference (Print vs Return)
def add_print(a, b):
    print(a + b)

def add_return(a, b):
    return a + b

x = add_print(2, 3)
y = add_return(2, 3)

print("x:", x)
print("y:", y)
```

```python
# Real Example
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

data = [10, 20, 30, 40, 50]
avg = calculate_average(data)

print("Average:", avg)
```

```python
# Find Maximum Value (Without Built-in max)
def find_max(numbers):
    maximum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num

    return maximum

data = [5, 12, 7, 20, 3]
print("Max value:", find_max(data))
```

```python
# Data Cleaning Example (remove negative value)
def remove_negative(numbers):
    clean_data = []

    for num in numbers:
        if num >= 0:
            clean_data.append(num)

    return clean_data

data = [10, -5, 20, -2, 30]
cleaned = remove_negative(data)

print("Cleaned data:", cleaned)
```

```python
# Positional Arguments
def info(name, age):
    print(name, age)

info("Sudip", 22)
```

```python
# Keyword Arguments
def info(name, age):
    print(name, age)

info(age=22, name="Sudip")
```

```python
# Default Arguments
def greet(name="Guest"):
    print("Hello", name)

greet()
greet("Sudip")
```

```python
# Multiple Return Values
def calculate(a, b):
    return a + b, a * b

sum_val, mul_val = calculate(4, 5)

print("Sum:", sum_val)
print("Multiply:", mul_val)
```

```python
# Function Returning Boolean (Very Useful)
def is_even(num):
    return num % 2 == 0

print(is_even(4))  # True
print(is_even(7))  # False
```

```python
# Student Data Analysis System
def calculate_average(marks):
    return sum(marks) / len(marks)

def find_grade(avg):
    if avg >= 80:
        return "A"
    elif avg >= 60:
        return "B"
    else:
        return "C"

def student_report(name, marks):
    avg = calculate_average(marks)
    grade = find_grade(avg)

    return name, avg, grade

name, avg, grade = student_report("Sudip", [70, 80, 90])

print("Name:", name)
print("Average:", avg)
print("Grade:", grade)
```
