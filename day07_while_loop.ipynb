```python
#  While Loop in Python
# A while loop runs as long as the condition is True

count = 0

while count < 5:
    print("Count:", count)
    count += 1   # important to avoid infinite loop
```

```python
# Infinite loop example

i = 0

while True:
    print(i)
    i += 1

    if i == 5:
        break   # stops the loop
```

```python
# Loop runs until a condition is met

number = 1

while number <= 5:
    print("Number:", number)
    number += 1
```

```python
# Access list using while loop (important concept)

data = [10, 20, 30, 40]

index = 0

while index < len(data):
    print("Value:", data[index])
    index += 1
```

```python
# Calculate total sum

numbers = [10, 20, 30, 40]

total = 0
i = 0

while i < len(numbers):
    total += numbers[i]
    i += 1

print("Total:", total)
```

```python
# Find max value using while

numbers = [12, 45, 7, 89, 23]

max_value = numbers[0]
i = 0

while i < len(numbers):
    if numbers[i] > max_value:
        max_value = numbers[i]
    i += 1

print("Maximum:", max_value)
```

```python
# Filter values greater than 25

data = [10, 30, 20, 50, 15]

filtered = []
i = 0

while i < len(data):
    if data[i] > 25:
        filtered.append(data[i])
    i += 1

print("Filtered data:", filtered)
```

```python
# Count how many values are greater than 20

data = [10, 25, 30, 5, 18]

count = 0
i = 0

while i < len(data):
    if data[i] > 20:
        count += 1
    i += 1

print("Count:", count)
```

```python
# Clean dataset (remove None values)

data = [10, None, 20, None, 30]

clean_data = []
i = 0

while i < len(data):
    if data[i] is not None:
        clean_data.append(data[i])
    i += 1

print("Clean data:", clean_data)
```

```python
# Skip specific value

i = 0

while i < 5:
    i += 1

    if i == 3:
        continue

    print(i)
```

```python
# Stop when value exceeds threshold

data = [10, 20, 30, 80, 40]

i = 0

while i < len(data):
    if data[i] > 50:
        print("Threshold exceeded at:", data[i])
        break
    i += 1
```

```python
# Simulate repeated process until condition met

attempts = 0

while attempts < 3:
    print("Attempt:", attempts + 1)
    attempts += 1

print("Process finished")
```
