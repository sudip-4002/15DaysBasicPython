# Day 9: Functions in Python

```python

# A function is a reusable block of code

def greet():
    print("Hello, welcome to Python!")

greet()
```

```python
# Function with input parameters

def greet_user(name):
    print("Hello,", name)

greet_user("Sudip")
```

```python
# Returning values from function

def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Result:", result)
```

```python
# Default argument values

def greet(name="Guest"):
    print("Hello,", name)

greet()
greet("Ram")
```

```python
# Returning multiple values

def calculate(a, b):
    return a + b, a - b

sum_val, diff_val = calculate(10, 5)

print("Sum:", sum_val)
print("Difference:", diff_val)
```

```python
# Passing arguments using keywords

def student_info(name, age):
    print("Name:", name)
    print("Age:", age)

student_info(age=22, name="Sudip")
```

```python
# *args allows multiple values

def total_sum(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(total_sum(10, 20, 30))
```

```python
# **kwargs for dictionary-like arguments

def user_details(**info):
    for key, value in info.items():
        print(key, ":", value)

user_details(name="Sudip", age=22, country="Nepal")
```

```python
# Anonymous function (short functions)

square = lambda x: x**2

print(square(5))
```

```python
# Nested function

def outer():
    def inner():
        print("Inner function")
    inner()

outer()
```

```python
# Remove missing and invalid values

def clean_data(data):
    clean = []

    for value in data:
        if value is None:
            continue
        if value < 0:
            continue
        clean.append(value)

    return clean

data = [10, None, -5, 20, 30]

print("Clean Data:", clean_data(data))
```

```python
# Average calculation (important in data analysis)

def calculate_average(data):
    total = 0

    for value in data:
        total += value

    return total / len(data)

numbers = [10, 20, 30, 40]

print("Average:", calculate_average(numbers))
```

```python
# Identify outliers

def detect_outliers(data, threshold):
    outliers = []

    for value in data:
        if value > threshold:
            outliers.append(value)

    return outliers

data = [10, 20, 100, 30, 200]

print("Outliers:", detect_outliers(data, 50))
```

```python
# Create new feature (square values)

def create_feature(data):
    result = []

    for value in data:
        result.append(value**2)

    return result

data = [2, 3, 4]

print("New Feature:", create_feature(data))
```

```python
# Categorize values

def categorize(data):
    low, medium, high = [], [], []

    for value in data:
        if value < 20:
            low.append(value)
        elif value < 50:
            medium.append(value)
        else:
            high.append(value)

    return low, medium, high

data = [10, 25, 60, 15, 45]

low, medium, high = categorize(data)

print("Low:", low)
print("Medium:", medium)
print("High:", high)
```

```python
# Function Declaration + Definition
# (In Python, declaration and definition happen together)

def calculate_sum(a, b):   # function name + parameters → declaration
    result = a + b         # logic → definition
    return result

# Function Call
output = calculate_sum(10, 5)

print("Sum is:", output)
```
