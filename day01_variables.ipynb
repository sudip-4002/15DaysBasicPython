```python
# Basic Variables
name = "Sudip"       # String
age = 21             # Integer
height = 5.7         # Float
is_student = True    # Boolean

# Display the values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student?", is_student)
```

```python
# Type checking
# Checking types of variables
print("\nType Checking Example:")
print("Type of name:", type(name))
print("Type of age:", type(age))
print("Type of height:", type(height))
print("Type of is_student:", type(is_student))
```

```python
# Multiple assignment
# Multiple assignment
x, y, z = 10, 20, 30
print("\nMultiple Assignment Example:")
print("x =", x, "y =", y, "z =", z)
```

```python
 # Type Conversion
num_str = "100"
num_int = int(num_str)
print("\nType Conversion Example:")
print("Original string:", num_str, "| Type:", type(num_str))
print("Converted to int:", num_int, "| Type:", type(num_int))
```

```python
# Constants (uppercase by convention)
PI = 3.14159
GRAVITY = 9.8
print("\nConstants Example:")
print("PI =", PI)
print("GRAVITY =", GRAVITY)
```
