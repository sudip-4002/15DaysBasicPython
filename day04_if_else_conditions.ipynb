```python
# Basic if condition
age = 18

if age >= 18:
    print("You are eligible to vote")
```

```python
# 2. If Else Condition

age = 16

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

```python
# 3. If Elif Else (Multiple Conditions)

marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

```python
# 4. Even or Odd Number

num = 7

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
# 5. Simple Login System

username = "admin"
password = "1234"

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 6. Nested If Condition

age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Allowed to enter")
    else:
        print("ID required")
else:
    print("Not allowed")
```
