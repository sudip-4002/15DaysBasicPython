```python
# Input from user
name = input("Enter your name: ")
age = input("Enter your age: ")

# Output to user
print("\nHello,", name)
print("You are", age, "years old")
```

```python
# Type casting example
age = input("Enter your age: ")  # input is string
print("Data type of age:", type(age))

# Convert string to integer
age_int = int(age)
print("Data type after conversion to int:", type(age_int))

# Convert integer to float
age_float = float(age_int)
print("Data type after conversion to float:", type(age_float))
```

Multiple type casting

```python
# Multiple type casting
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")

# Convert to float for calculation
num1 = float(num1)
num2 = float(num2)

# Sum of two numbers
total = num1 + num2
print("\nSum of", num1, "and", num2, "is:", total)
```
