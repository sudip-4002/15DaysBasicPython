#  File Handling



```python
# Creating a Text File

file = open("notes.txt", "w")

file.write("Welcome to Python File Handling")

file.close()
```

```python
# Reading a File

file = open("notes.txt", "r")

data = file.read()

print(data)

file.close()
```

```python
# Writing Student Data to File

file = open("students.txt", "w")

file.write("Ram\n")
file.write("Hari\n")
file.write("Sita\n")

file.close()
```

```python
# Reading Student Data

file = open("students.txt", "r")

print(file.read())

file.close()
```

```python
# Appending New Data

file = open("students.txt", "a")

file.write("Gita\n")

file.close()
```

```python
# Reading Updated File

file = open("students.txt", "r")

print(file.read())

file.close()
```

```python
# Reading File Line by Line

file = open("students.txt", "r")

for line in file:
    print(line)

file.close()
```

```python
# Using readlines()

file = open("students.txt", "r")

lines = file.readlines()

print(lines)

file.close()
```

```python
# Writing Employee Salary Data

file = open("salary.txt", "w")

file.write("Ram - 50000\n")
file.write("Hari - 45000\n")

file.close()
```

```python
# Reading Employee Salary Data

file = open("salary.txt", "r")

print(file.read())

file.close()
```

```python
# Counting Total Characters in File

file = open("notes.txt", "r")

data = file.read()

print(len(data))

file.close()
```

```python
# Counting Total Words in File

file = open("notes.txt", "r")

data = file.read()

words = data.split()

print(len(words))

file.close()
```

```python
# Checking File Exists

import os

if os.path.exists("notes.txt"):
    print("File Exists")
else:
    print("File Not Found")
```

```python
# Deleting a File

import os

if os.path.exists("temp.txt"):
    os.remove("temp.txt")
    print("File Deleted")
else:
    print("File Not Found")
```

```python
# Using with Statement

with open("notes.txt", "r") as file:
    print(file.read())
```

```python
# Writing Product Data

with open("products.txt", "w") as file:
    file.write("Laptop - 75000\n")
    file.write("Mouse - 1200\n")
```

```python
# Reading Product Data

with open("products.txt", "r") as file:
    print(file.read())
```

```python
# Creating User Feedback File

with open("feedback.txt", "w") as file:
    file.write("Good Service\n")
    file.write("Fast Delivery\n")
```

```python
# Reading User Feedback

with open("feedback.txt", "r") as file:
    print(file.read())
```

 Error Handling

```python
# Basic Try Except

try:
    number = 10 / 0
    print(number)

except:
    print("Error Occurred")
```

```python
# Handling ZeroDivisionError

try:
    result = 5 / 0

except ZeroDivisionError:
    print("Cannot Divide by Zero")
```

```python
# Handling ValueError

try:
    age = int(input("Enter Age: "))

    print(age)

except ValueError:
    print("Please Enter Numbers Only")
```

```python
# Using else Block

try:
    number = int(input("Enter Number: "))

    print(number)

except ValueError:
    print("Invalid Input")

else:
    print("Program Executed Successfully")
```

```python
# Using finally Block

try:
    result = 10 / 2

    print(result)

except:
    print("Error")

finally:
    print("Program Finished")
```

```python
# Handling File Error

try:
    file = open("data.txt", "r")

    print(file.read())

except FileNotFoundError:
    print("File Not Found")
```

```python
# Multiple Exceptions

try:
    number = int(input("Enter Number: "))

    result = 10 / number

except ValueError:
    print("Invalid Number")

except ZeroDivisionError:
    print("Cannot Divide by Zero")
```

```python
# Username Validation

try:
    username = input("Enter Username: ")

    if username == "":
        raise ValueError

    print(username)

except ValueError:
    print("Username Cannot Be Empty")
```

```python
# ATM Balance Check

balance = 5000

try:
    amount = int(input("Enter Withdraw Amount: "))

    if amount > balance:
        raise ValueError

    print("Withdraw Successful")

except ValueError:
    print("Insufficient Balance")
```

```python
# Student Marks Validation

try:
    marks = int(input("Enter Marks: "))

    if marks < 0 or marks > 100:
        raise ValueError

    print("Valid Marks")

except ValueError:
    print("Invalid Marks")
```

```python
# Login System Example

correct_password = "admin123"

try:
    password = input("Enter Password: ")

    if password != correct_password:
        raise ValueError

    print("Login Successful")

except ValueError:
    print("Wrong Password")
```
