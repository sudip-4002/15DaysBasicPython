```python
# Loops in Python
# A loop is used to repeat a block of code multiple times

# for loop example
for i in range(5):
    print("Iteration:", i)
```

```python
# range(start, stop, step)

for i in range(1, 10, 2):
    print(i)  # prints odd numbers from 1 to 9
```

```python
# Looping through a list

numbers = [10, 20, 30, 40]

for num in numbers:
    print("Value:", num)
```

```python
# Loop through characters in a string

name = "sudip"

for char in name:
    print(char)
```

```python
# while loop runs until condition becomes False

count = 0

while count < 5:
    print("Count:", count)
    count += 1  # important to avoid infinite loop
```

```python
# break stops the loop completely

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# continue skips current iteration

for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# pass does nothing (placeholder)

for i in range(3):
    pass  # used when loop is required but no action yet
```

```python
# Loop inside another loop

for i in range(3):
    for j in range(2):
        print("i:", i, "j:", j)
```

```python
# Data filtering example

numbers = [5, 12, 7, 20, 3]

for num in numbers:
    if num > 10:
        print("Greater than 10:", num)
```

```python
# Calculate total sum

numbers = [10, 20, 30, 40]

total = 0

for num in numbers:
    total += num

print("Total:", total)
```

```python
# Count how many numbers are greater than 15

numbers = [10, 25, 30, 5, 18]

count = 0

for num in numbers:
    if num > 15:
        count += 1

print("Count:", count)
```

```python
# Short way to create list using loop

numbers = [1, 2, 3, 4, 5]

squared = [x**2 for x in numbers]

print(squared)
```

```python
# Filtering even numbers from dataset

data = [1, 2, 3, 4, 5, 6]

even_numbers = []

for num in data:
    if num % 2 == 0:
        even_numbers.append(num)

print("Even numbers:", even_numbers)
```

Example using loops

```python
# Find the maximum value from a dataset

numbers = [12, 45, 7, 89, 23]

max_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num

print("Maximum value:", max_value)
```

```python
# Find the minimum value

numbers = [12, 45, 7, 89, 23]

min_value = numbers[0]

for num in numbers:
    if num < min_value:
        min_value = num

print("Minimum value:", min_value)
```

```python
# Calculate average of numbers

numbers = [10, 20, 30, 40]

total = 0

for num in numbers:
    total += num

average = total / len(numbers)

print("Average:", average)
```

```python
# Remove None values (missing data)

data = [10, None, 20, None, 30]

clean_data = []

for value in data:
    if value is not None:
        clean_data.append(value)

print("Clean data:", clean_data)
```

```python
# Count frequency of each element

data = [1, 2, 2, 3, 3, 3]

frequency = {}

for item in data:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

print(frequency)
```

```python
# Get students who passed (marks >= 40)

marks = [35, 60, 45, 20, 80]

passed_students = []

for m in marks:
    if m >= 40:
        passed_students.append(m)

print("Passed:", passed_students)
```

```python
# Convert temperatures from Celsius to Fahrenheit

celsius = [0, 10, 20, 30]

fahrenheit = []

for temp in celsius:
    f = (temp * 9/5) + 32
    fahrenheit.append(f)

print(fahrenheit)
```

```python
# Normalize values between 0 and 1

data = [10, 20, 30, 40]

min_val = min(data)
max_val = max(data)

normalized = []

for x in data:
    value = (x - min_val) / (max_val - min_val)
    normalized.append(value)

print(normalized)
```

```python
# Detect values greater than threshold

data = [10, 12, 15, 100, 14, 13]

outliers = []

for x in data:
    if x > 50:
        outliers.append(x)

print("Outliers:", outliers)
```

```python
# Create new feature: square of values

data = [2, 3, 4]

squared = []

for x in data:
    squared.append(x**2)

print(squared)
```
