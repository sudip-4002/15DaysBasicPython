# Creating a List

```python
# Student Marks Dataset

student_marks = [70, 85, 90, 60, 75]

print(student_marks)
```

```python
# Calculating Average Marks

student_marks = [70, 85, 90, 60, 75]

average = sum(student_marks) / len(student_marks)

print("Average Marks:", average)
```

```python
# Finding Highest and Lowest Salary

employee_salary = [25000, 40000, 55000, 30000, 70000]

print("Highest Salary:", max(employee_salary))
print("Lowest Salary:", min(employee_salary))
```

```python
# Removing Invalid Negative Data

temperature_data = [25, 30, -5, 28, -2, 32]

clean_data = []

for temp in temperature_data:
    if temp >= 0:
        clean_data.append(temp)

print("Clean Data:", clean_data)
```

```python
# Removing Duplicate User IDs

user_ids = [101, 102, 101, 103, 104, 102]

unique_users = set(user_ids)

print(unique_users)
```

```python
# Finding Total Unique Website Visitors

visitors = ["Ram", "Hari", "Ram", "Sita", "Hari", "Gita"]

unique_visitors = set(visitors)

print("Unique Visitors:", unique_visitors)
print("Total Unique Visitors:", len(unique_visitors))
```

```python
# Product Price Analysis

product_prices = [1200, 1500, 999, 2000, 1750]

average_price = sum(product_prices) / len(product_prices)

print("Average Price:", average_price)
print("Highest Price:", max(product_prices))
print("Lowest Price:", min(product_prices))
```

```python
# Monthly Sales Analysis

monthly_sales = [15000, 18000, 22000, 17000, 25000]

total_sales = sum(monthly_sales)

print("Total Sales:", total_sales)
print("Highest Sales:", max(monthly_sales))
```

# Tuples

```python
# Fixed GPS Coordinates using Tuple

location = (27.7172, 85.3240)

print("Latitude:", location[0])
print("Longitude:", location[1])
```

```python
# RGB Color Representation using Tuple

red_color = (255, 0, 0)

print(red_color)
```

```python
# Finding Passed Students

marks = [35, 80, 90, 20, 60, 45]

passed_students = []

for mark in marks:
    if mark >= 40:
        passed_students.append(mark)

print(passed_students)
```

```python
# Finding Even Numbers from Dataset

numbers = [10, 15, 20, 25, 30, 35]

even_numbers = []

for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)

print(even_numbers)
```

```python
# Online Shopping Cart Total

cart_prices = [500, 1200, 750, 300]

total_bill = sum(cart_prices)

print("Total Bill:", total_bill)
```

```python
# Attendance System using Set

attendance = ["Ram", "Hari", "Ram", "Sita", "Hari"]

unique_attendance = set(attendance)

print(unique_attendance)
```

```python
# Student Report Analysis

student_marks = [70, 85, 90, 70, 95, 85]

unique_marks = list(set(student_marks))

average = sum(unique_marks) / len(unique_marks)

print("Unique Marks:", unique_marks)
print("Average:", average)
print("Highest:", max(unique_marks))
print("Lowest:", min(unique_marks))
```
