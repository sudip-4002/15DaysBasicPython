# Day 8: Break & Continue in Python

```python

# Stop loop when a condition is met

for i in range(10):
    if i == 5:
        print("Stopping loop at:", i)
        break
    print(i)
```

```python
# Skip specific value

for i in range(5):
    if i == 2:
        continue
    print(i)
```

```python
# Simulating data processing with an error

data = [10, 20, 30, -1, 40, 50]

for value in data:
    if value < 0:
        print("Error: Invalid data found:", value)
        break   # stop processing
    print("Processing:", value)
```

```python
# Data cleaning (very important in data science)

data = [10, None, 25, None, 40]

clean_data = []

for value in data:
    if value is None:
        continue   # skip missing values
    clean_data.append(value)

print("Clean data:", clean_data)
```

```python
# Ignore invalid transactions

transactions = [100, -50, 200, -10, 300]

valid_transactions = []

for t in transactions:
    if t < 0:
        continue   # skip invalid
    valid_transactions.append(t)

print("Valid transactions:", valid_transactions)
```

```python
# Stop when threshold exceeded

data = [10, 20, 30, 80, 40]

for value in data:
    if value > 50:
        print("Alert! Threshold exceeded:", value)
        break
    print("Safe value:", value)
```

```python
# Remove extreme values (outliers)

data = [10, 12, 500, 15, 18]

filtered = []

for x in data:
    if x > 100:
        continue   # skip outlier
    filtered.append(x)

print("Filtered data:", filtered)
```

```python
# Search for a value and stop when found

data = [5, 10, 15, 20, 25]

target = 15

for value in data:
    if value == target:
        print("Found target:", value)
        break
```

```python
# Filtering categories

items = ["apple", "banana", "junk_food", "orange"]

for item in items:
    if item == "junk_food":
        continue   # skip unhealthy item
    print("Allowed item:", item)
```

```python
# Complex filtering logic

data = [5, None, 15, 200, 25]

processed = []

for value in data:
    if value is None:
        continue   # skip missing

    if value > 100:
        print("Stopping due to abnormal value:", value)
        break

    processed.append(value)

print("Processed data:", processed)
```

```python
# Using break & continue in while loop

data = [10, None, 30, -5, 40]

i = 0

while i < len(data):
    value = data[i]

    if value is None:
        i += 1
        continue

    if value < 0:
        print("Invalid value found:", value)
        break

    print("Valid:", value)
    i += 1
```
