```python
# Arithmetic Operators
# Used for mathematical calculations (VERY important in data science)

a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
# Comparison Operators
# Used to compare values → result is always True or False

x = 5
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

```python
# Logical Operators
# Used in conditions (VERY important for ML & data filtering)

a = True
b = False

print("AND:", a and b)
print("OR:", a or b)
print("NOT:", not a)
```

```python
# Assignment Operators
# Used to assign and update values

x = 10

x += 5
print(x)

x -= 3
print(x)

x *= 2
print(x)

x /= 4
print(x)
```

```python
# Identity Operators
# Check if two variables refer to same object in memory

a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)
print(a is c)
print(a == c)
```

```python
# Membership Operators
# Used to check if value exists in a sequence (list, string, etc.)

numbers = [1, 2, 3, 4, 5]

print(3 in numbers)
print(10 not in numbers)
```

```python
# Bitwise Operators
# Work on binary numbers (advanced but useful later)

a = 5   # 0101
b = 3   # 0011

print(a & b)
print(a | b)
print(a ^ b)
```
