```python
# Dictionary:
# A dictionary is a data structure that stores data in key-value pairs.
```

```python
# Creating a Dictionary

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

print(student)
```

```python
# Accessing Dictionary Values

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

print(student["name"])
print(student["age"])
```

```python
# Updating Dictionary Values

student = {
    "name": "Sudip",
    "age": 22
}

student["age"] = 23

print(student)
```

```python
# Adding New Key and Value

student = {
    "name": "Sudip",
    "age": 22
}

student["faculty"] = "CSIT"

print(student)
```

```python
# Removing Data from Dictionary

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

del student["age"]

print(student)
```

```python
# Looping Through Dictionary

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

for key, value in student.items():
    print(key, ":", value)
```

```python
# Getting Only Keys

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

print(student.keys())
```

```python
# Getting Only Values

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

print(student.values())
```

```python
# Using get() Method

student = {
    "name": "Sudip",
    "age": 22
}

print(student.get("name"))
print(student.get("faculty"))
```

```python
# Real-World Example: Student Marks System

student_marks = {
    "Math": 85,
    "Science": 90,
    "English": 78
}

print(student_marks)

average = sum(student_marks.values()) / len(student_marks)

print("Average Marks:", average)
```

```python
# Real-World Example: Employee Information

employee = {
    "name": "Ram",
    "salary": 50000,
    "department": "IT"
}

print(employee)
```

```python
# Real-World Example: Product Details

product = {
    "name": "Laptop",
    "price": 75000,
    "stock": 15
}

print(product)
```

```python
# Real-World Example: Word Frequency Counter

sentence = "data science is science"

words = sentence.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)
```

```python
# Real-World Example: Student Attendance System

attendance = {
    "Ram": "Present",
    "Hari": "Absent",
    "Sita": "Present"
}

print(attendance)
```

```python
# Real-World Example: Online Shopping Cart

cart = {
    "Laptop": 75000,
    "Mouse": 1200,
    "Keyboard": 2500
}

total = sum(cart.values())

print("Total Bill:", total)
```

```python
# Checking Key Exists or Not

student = {
    "name": "Sudip",
    "age": 22
}

if "name" in student:
    print("Key Exists")
```

```python
# Nested Dictionary

students = {
    "student1": {
        "name": "Ram",
        "marks": 85
    },
    "student2": {
        "name": "Hari",
        "marks": 90
    }
}

print(students)
```

```python
# Accessing Nested Dictionary Values

students = {
    "student1": {
        "name": "Ram",
        "marks": 85
    }
}

print(students["student1"]["name"])
print(students["student1"]["marks"])
```

```python
# Dictionary Length

student = {
    "name": "Sudip",
    "age": 22,
    "faculty": "CSIT"
}

print(len(student))
```

```python
# Copying a Dictionary

student = {
    "name": "Sudip",
    "age": 22
}

new_student = student.copy()

print(new_student)
```
